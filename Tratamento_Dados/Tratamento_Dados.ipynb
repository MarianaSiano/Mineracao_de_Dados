```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (15, 15)

dados = pd.read_csv('../Arquivo/trabalho2_dados_grupo-08.csv', index_col="Name")
```

```python
dados #dados da forma de como veio
```

```python
for column in dados.columns:
    print("'" + column + "'", " - ", type(column))
```

```python
dados.drop(["Candidate ID", "Number of characters in Original Name", "Month of Birth", "Year of Birth", "Gender"], axis=1, inplace=True)
dados.duplicated()
dados.drop_duplicates(inplace=True) #Remove linhas duplicadas
```

```python
dados.plot() #tabelas de todos os dados
```

```python
states = sorted(set(dados['State (Location)'])) #Adiciona os estados disponiveis
list_states = list(dados['State (Location)']) #Transforma em uma lista os valores do 'State (Location)

dic = {}
for state in states:
    dic[state] = float(list_states.count(state)) #Dicionario {Estado: Quantidade de pessoas}
```

```python
#Removendo o ruido baseado nos estados, estados com quantidade maiores ou iguais a 5 serão mantidos
labels = []
sizes = []

for x, y in dic.items(): #For para montar o gráfico de estados e remove os ruidos da tabela
    if(y > 4):
        labels.append(x)
        sizes.append(y)
    else:
        dados.drop(dados.loc[dados['State (Location)']==x].index, inplace=True)
```

```python
#Plot pie
plt.pie(sizes, labels=labels)
plt.show()
```

```python
dados #Após a remoção dos ruidos de estados

dados_copy = dados.copy() #Make backup
```

```python
def media_total_da_coluna(nome_coluna):
    #dados.drop(dados.loc[dados[nome_coluna]=="MD"].index, inplace=True) #Remove todos os MDs
    media_total_conta = 0.0
    count = 0
    
    for i in dados[nome_coluna]:
        if(i != "MD" and i != 0.0):
            media_total_conta += float(i)
            count += 1
    return media_total_conta / count
```

```python
dados = dados_copy.copy() #Apenas um backup

colunas_para_tratar = [' English 1','English 2','English 3','English 4','Quantitative Ability 1','Quantitative Ability 2','Quantitative Ability 3','Quantitative Ability 4','Domain Skills 1','Domain Skills 2','Domain Test 3','Domain Test 4','Analytical Skills 1','Analytical Skills 2','Analytical Skills 3']

##### TRATAMENTO DOS DADOS FALTANTES E INCONSISTENTES, ANALISE MANUAL

for coluna in colunas_para_tratar:
    media_TOTAL_CONTA = round(media_total_da_coluna(coluna ), 2)
    dados.loc[dados[coluna]=="MD", coluna] = media_TOTAL_CONTA #Muda tudo que está MD para media na coluna

for coluna in colunas_para_tratar:
    dados[coluna] = dados[coluna].astype(float); #Converte para float
    dados.drop(dados.loc[dados[coluna]==0.0].index, inplace=True) #Remove todos os valores iguais a 0.0
```
